```python
import pandas as pd
```

## 사망자수

```python
death = pd.read_csv("./rawdata/사망원인_104항목__성_시도별_사망자수_20221221170939.csv", encoding="cp949")
```

```python
death.head()
```

```python
death["시도별"].unique()
```

```python
death["시도별"].replace({"제주도": "제주특별자치도"}, inplace=True)
```

```python
death["사망원인별(104항목)"].unique()
```

```python
death_temp = death[death["사망원인별(104항목)"]=="순환계통 질환 (I00-I99)"].groupby("시도별").sum(numeric_only=True)
death_temp
```

## 심장장애인수

```python
disability = pd.read_csv("./rawdata/시도별_장애유형별_성별_등록장애인수_20221221170118.csv", encoding="cp949", skiprows=1)
```

```python
disability.head()
```

```python
disability.columns
```

```python
disability[["시도별", "심장", "심장.1", "심장.2"]]
```

```python
disability_temp = disability[["시도별", "심장"]]
disability_temp.head()
```

## 결합

```python
data = pd.merge(left=death_temp, right=disability_temp, how="outer", on="시도별")
```

```python
data
```

```python
data.시도별.unique()
```

```python
data.drop([12, 18], axis=0, inplace=True)
```

```python
data.columns = ["지역", "사망자", "심장장애"]
```
